5. Xem biến số và biến phân loại

```python
def classify_columns(df, name):
    numerical = df.select_dtypes(include=['number']).columns
    categorical = df.select_dtypes(include=['object']).columns

    print(f"\n--- Phân loại biến trong {name} ---")
    print(f"\nBiến số ({len(numerical)}):")
    print(numerical)

    print(f"\nBiến phân loại ({len(categorical)}):")
    print(categorical)
```

```python
classify_columns(df_train, "df_train")
```

```python
classify_columns(df_test, "df_test")
```

- Hàm xem các giá trị duy nhất trong từng cột

```python
def column_unique(df, column_name):
    # Đếm số lượng giá trị duy nhất trong cột
    unique_count = df[column_name].nunique()
    
    # Hiển thị số lượng giá trị duy nhất
    print(f"Số lượng giá trị duy nhất trong cột '{column_name}': {unique_count}")
```

```python
column_unique(df_train, columns_to_check)
```

- Hàm xem phân phối các giá trị trong từng cột

```python
def column_unique_values_and_counts(df, column_name):
    # Hiển thị tất cả các giá trị trong cột
    print(f"\nPhân phối các giá trị trong cột '{column_name}':")
    print(df[column_name].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

columns = df_train.columns

# Duyệt qua từng cột và vẽ biểu đồ phù hợp
for col in columns:
    plt.figure(figsize=(10, 6))  # Kích thước của biểu đồ
    if df_train[col].dtype == 'object':  # Nếu là dữ liệu phân loại
        sns.countplot(x=col, data=df_train)
        plt.title(f'Phân phối giá trị của cột: {col}')
        plt.xticks(rotation=90)  # Xoay nhãn trục x nếu cần
    else:  # Nếu là dữ liệu số
        sns.histplot(df_train[col], kde=True)
        plt.title(f'Phân phối giá trị của cột: {col}')
    
    plt.show()
```
